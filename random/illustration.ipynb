```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.plot(1,1,'k^',markersize=12,label="Ship")
plt.plot([2,3],[3,2],'bx',markeredgewidth=3,markersize=15,label="Fishing Grounds")
plt.xlim(0.5,4)
plt.ylim(0.5,4)

t = np.linspace(0,1,100)
plt.plot(1+t,1+2*t,'g--',lw=3,label="Fishing Ground 1 Attraction")
plt.plot(1+2*t,1+t,'g--',lw=3,alpha=.4,label="Fishing Ground 2 Attraction")
plt.arrow(1,1,.4,.6,width=.05,color='r',label="Instantaneous Trajectory")

plt.legend(loc="best")
plt.title("Determintion of Trajectory from Attractive Forces")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```
