```python
import numpy as np
import matplotlib.pyplot as plt

# Define rotation function
def rotate(vector, angle):
    angle_rad = np.deg2rad(angle)
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
                                [np.sin(angle_rad), np.cos(angle_rad)]])
    return np.dot(rotation_matrix, vector)

# Define vector components
s_i = np.array([0, 1.5])  # s_i points straight up and is slightly shorter
d_ij = np.array([1, 0.5])  # d_ij is initially at 30 degrees to the right of s_i
d_ij_rotated = rotate(d_ij, 30)  # rotate d_ij by 30 degrees to the right

# Make d_ij shorter
d_ij_shortened = d_ij_rotated * 0.5  # Scale the vector by 0.5 to make it shorter

# Calculate the orthogonal projection of d_ij onto s_i
projection = np.dot(d_ij_shortened, s_i) / np.dot(s_i, s_i) * s_i  # projection of d_ij_shortened onto s_i
projection_orthogonal = d_ij_shortened - projection

# Rotate the projection by 180 degrees
projection_rotated = rotate(projection_orthogonal, 180)

# Define additional vector p_i
p_i = rotate([1, 0], 20) * 1.2  # Make it slightly shorter than s_i and point up and to the right at 20 degrees
p_i_rotated = rotate(p_i, 90)  # Rotate p_i 90 degrees to the left

# Define starting point
x_i = np.array([0, 0])

# Plot the vectors
plt.figure(figsize=(6, 6))  # Adjusted figure size
plt.quiver(*x_i, *s_i, color='r', angles='xy', scale_units='xy', scale=1, label='$s_i$')
plt.quiver(*x_i, *d_ij_shortened, color='b', angles='xy', scale_units='xy', scale=1, label='$d_{ij}$')
plt.quiver(*d_ij_shortened, *projection_rotated, color='g', angles='xy', scale_units='xy', scale=1, label='$r_{ij}$')
plt.quiver(*x_i, *p_i_rotated, color='m', angles='xy', scale_units='xy', scale=1, label='$p_i$')

# Plot the origin point triangle
plt.plot(*x_i, marker='^', markersize=12, color='black')

# Plot random points
points = [(0.1, 2.2), (0.26, 1.9), (-0.5, 1.3), (0.8, 1.2)]
for i, point in enumerate(points):
    if i == 0:
        plt.scatter(*point, color='darkgoldenrod', marker='x', label='Fishing Grounds')
    else:
        plt.scatter(*point, color='darkgoldenrod', marker='x')

# Plot annotations
plt.text(x_i[0] - 0.1, x_i[1] - 0.1, '$x_i$', fontsize=12, ha='right', va='bottom')  # Adjusted position for x_i label
plt.text(*x_i + s_i, '$s_i$', fontsize=12, ha='left', va='bottom')
plt.text(*x_i + d_ij_shortened, '$d_{ij}$', fontsize=12, ha='left', va='top')
plt.text(*x_i + d_ij_shortened + projection_rotated / 2, '$r_{ij}$', fontsize=12, ha='right', va='top')
plt.text(*x_i + p_i_rotated, '$p_i$', fontsize=12, ha='left', va='bottom')

# Set plot limits and labels
plt.xlim(-1, 1.25)  # Adjusted x-axis limits
plt.ylim(-0.5, 2.5)  # Adjusted y-axis limits
plt.xlabel('x')
plt.ylabel('y')
plt.title('Illustration of Lemma 2')

# Display legend
plt.legend()

# Show plot
plt.tight_layout()  # Adjust layout to reduce whitespace
plt.show()
```
